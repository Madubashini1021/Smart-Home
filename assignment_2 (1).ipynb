Outlier Detection and Removal

1. Interquartile Range (IQR) Method

```python
from scipy import stats

# IQR Method
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Z-score Method
def detect_outliers_zscore(df, column, threshold=3):
    z_scores = stats.zscore(df[column])
    outliers = df[(z_scores > threshold) | (z_scores < -threshold)]
    return outliers
```

```python
numeric_columns = [
    'use [kW]', 'gen [kW]', 'House overall [kW]', 'Dishwasher [kW]',
    'Furnace 1 [kW]', 'Furnace 2 [kW]', 'Home office [kW]', 'Fridge [kW]',
    'Wine cellar [kW]', 'Garage door [kW]', 'Kitchen 12 [kW]', 'Kitchen 14 [kW]',
    'Kitchen 38 [kW]', 'Barn [kW]', 'Well [kW]', 'Microwave [kW]', 'Living room [kW]',
    'Solar [kW]', 'temperature', 'humidity', 'visibility', 'apparentTemperature',
    'pressure', 'windSpeed', 'windBearing', 'precipIntensity', 'dewPoint', 'precipProbability'
]

# Detect outliers for each column
outliers_summary = {}

for column in numeric_columns:
    # IQR Method
    iqr_outliers = detect_outliers_iqr(df, column)
    iqr_outlier_count = iqr_outliers.shape[0]

    # Z-score Method
    zscore_outliers = detect_outliers_zscore(df, column)
    zscore_outlier_count = zscore_outliers.shape[0]

    # Store the results
    outliers_summary[column] = {
        "IQR Outliers": iqr_outlier_count,
        "Z-score Outliers": zscore_outlier_count
    }

# Print outliers summary
outliers_summary_df = pd.DataFrame(outliers_summary).T
print(outliers_summary_df)
```

```python
# Define a function to cap outliers based on IQR
def cap_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].apply(lambda x: upper_bound if x > upper_bound else x)
    return df

# Columns to apply capping
columns_to_cap = [
    'use [kW]', 'gen [kW]', 'House overall [kW]', 'Dishwasher [kW]',
    'Furnace 1 [kW]', 'Furnace 2 [kW]', 'Home office [kW]', 'Wine cellar [kW]',
    'Garage door [kW]', 'Kitchen 12 [kW]', 'Kitchen 14 [kW]', 'Living room [kW]',
    'Solar [kW]', 'Barn [kW]', 'Well [kW]', 'visibility', 'precipIntensity'
]

# Apply capping to selected columns
for column in columns_to_cap:
    df = cap_outliers_iqr(df, column)
```

To further test the hypothesis, categorize solar_power into low, medium, and high levels, then use ANOVA to see if there are statistically significant differences in total energy consumption across these categories.

```python
from scipy import stats
import numpy as np

# Create categories for solar power generation based on quantiles
df['Solar [kW]'] = pd.qcut(df['Solar [kW]'], q=3, labels=['low', 'medium', 'high'])

# Group total energy consumption by solar power level
low_solar = df[df['Solar [kW]'] == 'low']['use [kW]']
medium_solar = df[df['Solar [kW]'] == 'medium']['use [kW]']
high_solar = df[df['Solar [kW]'] == 'high']['use [kW]']

# Perform one-way ANOVA
anova_result = stats.f_oneway(low_solar, medium_solar, high_solar)

# Print the ANOVA results
print("\nANOVA Analysis:")
print("F-statistic:", anova_result.statistic)
print("p-value:", anova_result.pvalue)
```

Comparative Analysis: High vs. Low Usage

```python
# Import necessary libraries
import pandas as pd
import statsmodels.api as sm
import numpy as np

# Load data into a DataFrame (assuming 'df' is your DataFrame)
# df = pd.read_csv('your_data.csv')  # Uncomment and replace with your actual file

# Define independent (predictor) and dependent (response) variables
# Include specific high-energy appliances as predictors
X = df[['Dishwasher [kW]', 'Fridge [kW]', 'Furnace 1 [kW]', 'Furnace 1 [kW]']]  # Add other relevant appliances
y = df['use [kW]']  # Dependent variable: total energy consumption

# Handle missing or infinite values in X
# Replace infinite values with NaN
X.replace([np.inf, -np.inf], np.nan, inplace=True)
# Remove rows with any missing values in X or y
# This will ensure that both X and y have the same number of rows
X = X.dropna()
y = y[X.index]  # This line was moved up to maintain the order of operations

# Add a constant to the model (intercept)
X = sm.add_constant(X)

# Fit the multiple linear regression model
model = sm.OLS(y, X).fit()

# Print the regression summary
print("Multiple Linear Regression Analysis for Appliance Usage and Overall Energy Consumption:")
print(model.summary())
```
